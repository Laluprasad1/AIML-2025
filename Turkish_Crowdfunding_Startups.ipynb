**1. Identify the top-10 Startups in Turkey**

```python
df['toplanan_tutar'] = pd.to_numeric(df['toplanan_tutar'], errors='coerce')
top_10_startups = df.nlargest(10, 'toplanan_tutar')[['proje_adi', 'toplanan_tutar']]
top_10_startups
```

**2. Find the Max and Min crowdsourced funding obtained by Turkey**

```python
max_funding = df['toplanan_tutar'].max()
min_funding = df['toplanan_tutar'].min()
print("Maximum funding obtained:", max_funding)
print("Minimum funding obtained:", min_funding)
```

**3. Name the gender which has acquired the most funding for Startups**

```python
funding_by_gender = df.groupby('proje_sahibi_cinsiyet')['toplanan_tutar'].sum()
most_funded_gender = funding_by_gender.idxmax()
print("Gender with the most funding:", most_funded_gender)
```

**4. Identify the top-5 cities where startup is most**

```python
top_5_cities = df['konum'].value_counts().head(5)
print("Top 5 cities with the most startups:\n")
top_5_cities
```

**5. Name the Top-5 technologies popular in Turkish startups.**

```python
top_5_technologies = df['kategori'].value_counts().head(5)
print("Top 5 technologies/categories:\n")
top_5_technologies
```

**6. Apply either Classification or clustering or regression model and evaluate the accuracy, error metrics of the model**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = df.dropna(subset=['toplanan_tutar'])
X = df[['hedef_miktari']]
y = df['toplanan_tutar']
X = pd.to_numeric(X['hedef_miktari'], errors='coerce').fillna(0).values.reshape(-1, 1)
y = pd.to_numeric(y, errors='coerce').fillna(0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
